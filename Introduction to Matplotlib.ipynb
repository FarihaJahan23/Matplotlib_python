# Matplotlib example workflow:

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. Setup Plot
fig,ax=plt.subplots(figsize=(5,6))

# 3. plot data
ax.plot(x,y)

# 4. customize data
ax.set(title="Simple Plot",
      xlabel='x-axis',
      ylabel='y-axis')

# 5. save and show the whole figure
plt.savefig('figures/sample-plot.png')
```

# Scatter Plot and Bar Plot

```python
# Making figures with NumPy arrays,
#Create some data
x=np.linspace(0,10,100)
x[:10]
```

```python
#plot the data
fig,ax=plt.subplots(figsize=(5,4))
ax.plot(x,x**2)# this is called as line plot
```

```python
#use some data to make a scatter
fig,ax=plt.subplots(figsize=(5,4))
ax.scatter(x,np.exp(x))# creating an exponential graph
```

```python
#another scatter plot with sin function
fig,ax=plt.subplots(figsize=(5,4))
ax.scatter(x,np.sin(x))
```

```python
## Bar Plot
# Bar plot using python dictionary
#make a plot from dictionary

nut_butter_prices={"Almond Butter": 10,
"Peanut Butter": 8,
"Cashew Butter": 12}

fig,ax=plt.subplots(figsize=(5,4))
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title="Dan's Nut Butter Store",
      ylabel="Price ($)")
```

```python
## Horizontal bar creation,
fig,ax=plt.subplots(figsize=(5,4))
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))#It has to be turned into a list.
```

```python
# Plotting histogram,
#make some data
x=np.random.randn(1000)#The numpy.random.randn() function creates an array of specified shape and fills it with random values
fig,ax=plt.subplots(figsize=(5,4))
ax.hist(x)
```

```python
#Sub plotting systems

#subplot option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
#Sub plotting systems

#subplot option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each different axis
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
```

## Importing csv file and performing visualizations

```python
df=pd.read_csv('car-sales.csv')
```

```python
df
```

```python
# Replacing the expressions
df["Price"]=df["Price"].str.replace('$','')
```

```python
df
```

```python
df["Price"]=df["Price"].str[:-2]
df
```

```python
#Converting the price column to a numeric value. This example includes to keep the data with decimal points.
df["Price"]=df["Price"].str.replace('[\$\,]|\.\d*','').astype(float)
```

```python
# Adding date column
df["Sale Date"]=pd.date_range("1/1/2020",periods=len(df))
df
```

```python
#Appling Cumulative summation
df["Total Sales"]=df["Price"].cumsum().astype(float)
df
```

```python
#lets plot the total/cumsum sales
#line plot
df.plot(x="Sale Date",y="Total Sales")
plt.ylabel("Price ($)")
```

```python
#reassign price column to int
df["Price"]=df["Price"].astype(int)
#scatter plot
df.plot(x="Odometer (KM)", y="Price",kind="scatter")#this is called matplotlib OO method, used for advanced graphs
```

```python
# Car graph from the imported data frame,
df.plot(x="Make", y="Odometer (KM)", kind="bar")
```

```python
#histogram
df["Odometer (KM)"].plot.hist()
#df["Odometer (KM)"].hist() #another way
```

```python
df["Total Sales"].plot(kind="bar",color=["salmon"])
```

```python
over_50.plot(x='age',
             y='chol',
             kind='scatter',
             c='target')
```

```python
##OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
       y=over_50["chol"],
       c=over_50["target"])

#Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
      ylabel="Cholesterol")

#add a legend
ax.legend(*scatter.legend_elements(),title="Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--')
```

```python
##subplot of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))

#add data to ax0
scatter=ax0.scatter(x=over_50["age"],
       y=over_50["chol"],
       c=over_50["target"])

#Customize the ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
      ylabel="Cholesterol")

#add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

#add a horizontal line
ax0.axhline(over_50["chol"].mean(),linestyle='--')


#add data to ax1
scatter=ax1.scatter(x=over_50["age"],
       y=over_50["thalach"],
       c=over_50["target"])

#Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
      ylabel="Max Heart Rate")

#add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

#add a horizontal line
ax1.axhline(over_50["thalach"].mean(),linestyle='--')
```
